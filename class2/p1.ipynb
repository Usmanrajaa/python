```python
a=3
b=4
print(a & b)
print(a|b)
print(~a)
```

```python
## XOR Operator
# Original message
message = 5         # (binary 0101)
security_key = 3    # (binary 0011)

# Encryption
encrypted = message ^ security_key
print("Encrypted:", encrypted)

# Decryption (apply XOR again with same key)
decrypted = encrypted ^ security_key
print("Decrypted:", decrypted)
```
