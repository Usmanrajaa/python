## 1. Create

# Rules for Dictionary creation
#### RULES FOR KEYS
#### 1. Key could be immutable type ( numeric, string, float,boolean,tuple)
#### 2. key cann't be taken any mutable type like list,set,dictionary
#### 3. all keys should be unique. if duplicate keys comes then it overwrites earlier one.
#### 4. keys are case sensitive.

#### RULES FOR VALUES
#### 1.values could be taken anything i.e (numeric,string,float ,list ,tuple,set, dictionary...etc)
#### 2.Duplicate values could be taken for multiple different keys.

```python
d1={"name":"alex","profession":"SDE","NAME":"raj"}
d2={70:"title"}
d4={"product_count":[2,9]}
d5={"title":70}
d6={(2,3):"product_count"}
d1={"name":"alex","profession":"SDE","NAME":"raj"}

```
