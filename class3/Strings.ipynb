## Object cloning

```python
original1=["apple","banana","bread"]
original2=["apple",["banana","bread"]]

```

```python
shallow1=original1.copy()
shallow2=original2.copy()
```

```python
shallow1[1]="butter"
print(shallow1)
print(original1)
```

```python
shallow2[1][0]="butter"
print(shallow2)
print(original2)
```

```python
import copy
deep_original1=["apple","milk","bread"]
deep_original2=["apple",["milk","bread"]]
```

```python
deep_original1_copy=copy.deepcopy(deep_original1)
deep_original2_copy=copy.deepcopy(deep_original2)
```

```python
print(deep_original1)
print(deep_original1_copy)
```

```python
deep_original1_copy[1]="buiscuit"
```
