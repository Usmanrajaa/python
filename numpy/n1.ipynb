### 5.NumPy Array Iteration, Joining, and Splitting

```python
a1=np.array([1,2,3,5])
a2=np.array([[2,3,4],[5,6,7],[9,8,1]])
a3=np.array([[[1,2,3],[5,6,7],[6,7,7]]])
```

```python
a3.ndim
```

```python
for i in a1:
    print(i,end=" ")
```

```python
a2
```

```python
for i in a2:
    for j in i:
        print(j,end=" ")
```

```python
for i in a3:
    for j in i:
        for k in j:
            print(k,end=" ")
```

```python
for i in np.nditer(a3):
    print(i,end=" ")
```

```python
x2=np.vstack((l1,l2,l3))  ## np.stack=np.vstack
```

```python
print(x1)
print(x2)
print(x3)
```
