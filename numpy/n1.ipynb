# NumPy Random Module Cheat Sheet

---

# 1. Core Sampling Methods

```python
rng = np.random.default_rng()
```

| Method | Description |
|--------|-------------|
| `rng.random(size=None)` | Returns random floats uniformly in the interval `[0.0, 1.0)` |
| `rng.integers(low, high=None, size=None, endpoint=False)` | Returns random integers from `low` (inclusive) to `high` (exclusive/inclusive depending on `endpoint`) |
| `rng.choice(a, size=None, replace=True, p=None, axis=0)` | Random sample from a 1D array or integer range |
| `rng.bytes(length)` | Returns a random string of bytes |

---

# 2. Permutations and Shuffling

| Method | Description |
|--------|-------------|
| `rng.shuffle(x, axis=0)` | Shuffles an array **in-place** |
| `rng.permutation(x, axis=0)` | Returns a randomly permuted copy |
| `rng.permuted(x, axis=None)` | Permutes elements independently along an axis |

---

# 3. Statistical Distributions

> **All distribution methods accept a `size` parameter.**

| Method | Description |
|--------|-------------|
| `rng.standard_normal(size=None)` | Standard Normal distribution (μ=0, σ=1) |
| `rng.normal(loc=0.0, scale=1.0, size=None)` | Normal (Gaussian) distribution |
| `rng.uniform(low=0.0, high=1.0, size=None)` | Uniform distribution |
| `rng.binomial(n, p, size=None)` | Binomial distribution |
| `rng.poisson(lam=1.0, size=None)` | Poisson distribution |
| `rng.exponential(scale=1.0, size=None)` | Exponential distribution |
| `rng.beta(a, b, size=None)` | Beta distribution |
| `rng.gamma(shape, scale=1.0, size=None)` | Gamma distribution |
| `rng.chisquare(df, size=None)` | Chi-square distribution |
| `rng.standard_t(df, size=None)` | Student's t distribution |
| `rng.f(dfnum, dfden, size=None)` | F (Fisher-Snedecor) distribution |
| `rng.logistic(loc=0.0, scale=1.0, size=None)` | Logistic distribution |
| `rng.lognormal(mean=0.0, sigma=1.0, size=None)` | Log-normal distribution |
| `rng.multinomial(n, pvals, size=None)` | Multinomial distribution |
| `rng.multivariate_normal(mean, cov, size=None)` | Multivariate Normal distribution |
| `rng.dirichlet(alpha, size=None)` | Dirichlet distribution |
| `rng.geometric(p, size=None)` | Geometric distribution |
| `rng.gumbel(loc=0.0, scale=1.0, size=None)` | Gumbel distribution |
| `rng.hypergeometric(ngood, nbad, nsample, size=None)` | Hypergeometric distribution |
| `rng.laplace(loc=0.0, scale=1.0, size=None)` | Laplace (Double Exponential) distribution |
| `rng.logseries(p, size=None)` | Logarithmic Series distribution |
| `rng.negative_binomial(n, p, size=None)` | Negative Binomial distribution |
| `rng.noncentral_chisquare(df, nonc, size=None)` | Non-central Chi-square distribution |
| `rng.noncentral_f(dfnum, dfden, nonc, size=None)` | Non-central F distribution |
| `rng.pareto(a, size=None)` | Pareto distribution |
| `rng.power(a, size=None)` | Power distribution |
| `rng.rayleigh(scale=1.0, size=None)` | Rayleigh distribution |
| `rng.standard_cauchy(size=None)` | Standard Cauchy distribution |
| `rng.standard_exponential(size=None)` | Standard Exponential distribution |
| `rng.standard_gamma(shape, size=None)` | Standard Gamma distribution |
| `rng.triangular(left, mode, right, size=None)` | Triangular distribution |
| `rng.vonmises(mu, kappa, size=None)` | Von Mises distribution |
| `rng.wald(mean, scale, size=None)` | Wald (Inverse Gaussian) distribution |
| `rng.weibull(a, size=None)` | Weibull distribution |
| `rng.zipf(a, size=None)` | Zipf distribution |

# 4. Legacy API (`np.random.*`)

> ⚠️ These functions use the older **Mersenne Twister** generator.
>
> **Use `default_rng()` for new code.**

---

## Simple Random Data Functions

| Method | Description |
|--------|-------------|
| `np.random.rand(d0, d1, ..., dn)` | Uniform floats in `[0,1)` |
| `np.random.randn(d0, d1, ..., dn)` | Standard Normal distribution |
| `np.random.randint(low, high=None, size=None)` | Random integers |
| `np.random.random_sample(size=None)` | Uniform floats |
| `np.random.random(size=None)` | Alias of `random_sample()` |
| `np.random.ranf(size=None)` | Alias of `random_sample()` |
| `np.random.sample(size=None)` | Alias of `random_sample()` |
| `np.random.choice(a, size=None, replace=True, p=None)` | Random selection |
| `np.random.bytes(length)` | Random bytes |

---

## Shuffling and Seeding

| Method | Description |
|--------|-------------|
| `np.random.shuffle(x)` | Shuffle array in-place |
| `np.random.permutation(x)` | Return shuffled copy |
| `np.random.seed(seed=None)` | Set global random seed |
| `np.random.get_state()` | Get generator state |
| `np.random.set_state()` | Restore generator state |

---

## Distribution Functions

The legacy API provides the same distribution names:

```python
np.random.normal()
np.random.uniform()
np.random.binomial()
np.random.poisson()
np.random.exponential()
...
```

# 5. Helper & Bit Generator Infrastructure

| Method | Description |
|--------|-------------|
| `np.random.default_rng()` | Creates the recommended Generator |
| `np.random.RandomState()` | Legacy random generator |
| `np.random.PCG64` | High-quality bit generator |
| `np.random.PCG64DXSM` | Improved PCG64 variant |
| `np.random.MT19937` | Mersenne Twister bit generator |
| `np.random.Philox` | Counter-based generator |
| `np.random.SFC64` | Fast Small Counter generator |
| `np.random.SeedSequence` | Advanced seed management |

---

✅ **Use this in new code**

```python
import numpy as np

rng = np.random.default_rng(42)

numbers = rng.normal(size=10)
```

```python
np.random.seed(42)
np.random.normal(size=10)
```

### 5.NumPy Array Iteration, Joining, and Splitting

````python
a1=np.array([1,2,3,5])
a2=np.array([[2,3,4],[5,6,7],[9,8,1]])
a3=np.array([[[1,2,3],[5,6,7],[6,7,7]]])
````

````python
a3.ndim
````

````python
for i in a1:
    print(i,end=" ")
````

````python
a2
````

````python
for i in a2:
    for j in i:
        print(j,end=" ")
````

````python
for i in a3:
    for j in i:
        for k in j:
            print(k,end=" ")
````
