#### method2. From a List of Dictionaries

```python
data=[
    {'name':'a','age':20,'score':60,'subject':"math"},
    {'name':'b','age':30,'score':70,'subject':"physcis"},
    {'name':'c','age':40,'score':80,'subject':"chemistry"},
    {'name':'d','age':50,'score':90,'subject':"bio"}
]
```
