```python
is_tired=False
i=1
is_coach_available=True
coach_list=["raman","tanuj"]
coach_checker=False

while not is_tired and i<10 and is_coach_available and not coach_checker:  
    print(f"round {i} is successfully done")
    for j in coach_list:
        if j=="hardik":
            coach_checker=True
    i=i+1
print("all round running is over")
```
