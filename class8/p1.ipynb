```python
class Bankbalance:
    def __init__(self,owner,__balance):
        self.owner=owner
        self.__balance=__balance ## private
    def get(self):
        print(f"the total balance remainint is {self.__balance}")
    def set(self,amount):
        self.amount=amount
        if self.amount>0:
            self.__balance=self.__balance+amount
        else:
            print("given amount is null")

bank=Bankbalance("aman",9000)
bank.set(8000)
bank.get()
```

## 2. INHERITANCE

```python
class Bus:
    def __init__(self,brand,color,model):
        self.brand=brand
        self.color=color
        self.model=model
    def show(self):
        print(f"i have bus of this brand {self.brand} and color {self.color} and model {self.model}")
class electric_bus(Bus):
    def __init__(self,brand,color,model,battery_capacity):
        super(). __init__(brand,color,model)
        self.battery_capacity=battery_capacity
        
    def show(self):
        print(f"i have electric bus of color {self.color} and brand{self.brand} and model{self.model} battery_capacity {self.battery_capacity}")

class cng_bus(electric_bus,Bus):
    def __init__(self,brand,color,model,mileage,battery_capacity):
        super(). __init__(brand,color,model,battery_capacity)
        self.mileage=mileage   
    def show(self):
        print(f"i have cng_bus of color {self.color} and brand{self.brand}")


b1=Bus("volvo","red",2017)
b1.show()
b2=electric_bus("tata","blue",2018,"5000mAH")
b2.show()

```

## 3. Polymorphism

```python
class Animal:
    def sound(self):
        return "make some generic sound"
class Dog(Animal): ## inheritance
    def sound(self): ## overriding
        return " dog sounds is Bark"
class Cat(Animal):
    def sound(self): ## overriding
        return "cat sound is Meow"
class Lion(Animal):
    def sound(self): ## overriding
        return "lions sound is roar"
a=[Animal(),Dog(),Cat(),Lion()]
for i in a:
    print(i.sound())
```

## 4. Abstraction

```python
## without abstraction
class Food_stall:
    def ask_menu(self):
        print("give me your restaurant menu")
    def give_order(self):
        print("i want to order coffe and tea")
    def cooking_start(self):
        print("cook has started preparing your order")
    def deliver_food(self):
        print("your order has been delivered at your table")
a=Food_stall()
a.ask_menu()
a.give_order()
a.cooking_start()
a.deliver_food()

```

```python
from abc import ABC,abstractmethod
class order(ABC):
    @abstractmethod
    def order_food(self):
        pass

class food_order(order):
    def order_food(self):
        print("give me your restaurant menu")
        print("i want to order coffe and tea")
        print("cook has started preparing your order")
        print("your order has been delivered at your table")
a=food_order()
a.order_food()

```
