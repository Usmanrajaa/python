```python
#3 dependent conditional statement
x=40
y=90
z=150
if x<20:
    print("test 1 correct")
elif y>100:
    print("test 2 correct")
elif z>200:
    print("test 3 correct")
else:
    print("No record found")
```

```python
#4. nested conditinal statement
x=40
y=90
z=150
if x>20:
    if y>100:
        if z>200:
            print("all criteria passed and this is correct candidate")
        else:
            print("failed at 3rd condtion")
    else:
        print("failed at 2nd condition")
else:
    print("failed at 1st condtion")
```
