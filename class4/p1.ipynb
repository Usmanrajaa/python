## Decision Control
### 1. if-else

```python
a1=50
b1=60
c1=80

if b1>c1:
    print("correct")
else:
    print("incorrect")
```

```python
a1=50
b1=60
c1=80

if b1>c1 and a1>c1:
    print("correct")
else:
    print("incorrect")
```

#### multiple nested logical testing

```python
a1=107
b1=90
if a1>b1:
    if a1-b1>=10:
        if (a1-b1)%2==0:
            print("it is best choice because difference is even")
        else:
            print("could choose but not good choice because difference is odd")

    else:
        print("it is correct but not choosing this because difference is less than 10")
else:
    print("incorrect")
```

### 2. Multiple condition with elif

```python
a1=50
b1=60
c1=80
## Q1. if a<b and b<c then print "a<c". otherwise print any fallback message.
if a1>b1 and b1<c1:
    print("pqr")
elif a1>c1 and b1>a1:
    print("xyz")
elif a1<b1 and a1<c1:
    print("mno")
else:
    print("not matching the criteria")

```

### 3. Nested if

```python
## take a user input number (>=0) and check if it is "0" then print "zero" otherwise print "positive number". 
## if not fall under any of these then  print "negative number"
```

```python
a=int(input("enter a number:"))
if a==0:
    print("zero")
elif a>0:
    print("positive number")
else:
    print("negative number")
```

```python
x=int(input("enter randon number"))
if x>=0:
    if x==0:
        print("zero")
    else:
        print("positive number")
else:
    print("negative number")   

```

```python
## Q2.1. take a randonm number and check if number<100 then say it "upto 100",if number>100 but <150 then say it "100-150 range". 
## if number>150 but less than 200 then say it "200-150 range"
## Q2.2 for the same random number if number<100 then print check divisible by 2 and print "even no." else "odd no." otherwise not less than 100.
```

```python
z=int(input("Enter random number:"))
if z<=100:
    print("upto 100")
elif z>=100 and z<=150:
    print("100-150 range")
elif z>=150 and z<=200:
    print("150-200 range")
else:
    print("200+")

if z<=100:
    if z%2==0:
        print("even no.")
    else:
        print("odd no.")
else:
    print("not less than 100")
    
```
